```python
import os
import glob

# set the parent directory to the present directory
basedir = os.getcwd() 
# or comment out the line above above and uncomment the following 
# two lines to set the directory directly as a file-system appropriate string
# basedir = '/whatever/you/want'
# os.chdir(basedir)

pdfdir = basedir+'/pdf'
htmldir = basedir+'/html'

# get an array of notebook filenames
nbs = glob.glob("*.ipynb")

for nb in nbs:
    command='jupyter nbconvert '+nb+' --to pdf --output-dir '+pdfdir
    os.system(command)
    command='jupyter nbconvert '+nb+' --to html --output-dir '+htmldir
    os.system(command)
```
